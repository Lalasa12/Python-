# Type of variables


# 1.Instance Variables

- Instance variables belong to individual objects.
- Each Object has its own copy.
- They are created by using self.variable_name.



# ways to declare instance variables.

1. Inside the Constructor (__init__) using self.

2. Inside the instance method using self.

```python
# inside the constructor
class Test:
    def __init__(self):
        self.a = 10
        self.b = 20

# object creation

t = Test()
        
```

```python
t.a
```

```python
t.b
```

```python
print(t.__dict__)   # it shows the all instance variables of an object
```

```python
# inside the instance method
class Test:
    def __init__(self):
        self.a = 10
    def m1(self):
        self.b = 20
        self.c = 30
        return self.b,self.c

t1 = Test()
```

```python
t1.m1()
```

```python
t1.__dict__
```

```python
# outside the class using object reference
class Test:
    def __init__(self):
        self.a = 10
t = Test()
t.b = 20    # added outside the class 
```

```python
t.a
```

```python
t.b
```

```python
t.__dict__
```

# Note

- t.b = 20 directly adds a new instance variable to the object t.

- No self is needed outside the class.

# 2.Class Variables

- Class variable belongs to only class not the objects.
- one copy only,shared the all objects.
- created once at class level.
- defined outside the constructor and methods but inside the class.

```python
class Employee:
    Bankname = "SBI"     # class variable
    def __init__(self,id,name,dept):
        self.eid = id
        self.ename = name
        self.dept = dept
    def s1(self):
        self.salary = float(input())
        return f"The employee id is {self.eid}, name is {self.ename} and salary is {self.salary}"
T = Employee(45,"LalasaReddy","HR")
T1 = Employee(55,"Chitti","IT")
```

```python
# Inside the classmethod using cls or classname

class Test:
    @classmethod
    def m2(cls):
        cls.a = 100  # using cls(the class itself)
        Test.b = 200 # using the class name directly)
t1 = Test()
```

```python
t1.m2()
```

```python
Test.a
```

```python
Test.b
```

# Note


- 1. @classmethod - A classmethod receives the class itself(cls) as the first arguement,instead of an instance(self).This means you can use cls to access or modify class level attributes.
  2. cls.a = 100,Here cls refers to the class Test.So this line creates a class attribute a with value 100.Equivalent to Test.a = 100.
  3. Test.b = 200 this directly sets another class attribute b with the value 200.This is same as cls.b = 200.

# 1. Instance Methods

- It is a user defined function which is defined inside a class that works with instance variables.
- The first parameter of instance method is self(pointer).
- It is called object reference.methodname().

```python
class student:
    def __init__(self,rollno,name,marks):
        self.rollno = rollno
        self.name = name
        self.marks = marks
    def display(self):
        return f" Rollno:{self.rollno}, name : {self.name} , marks : {self.marks}"
    def grade(self):
        if self.marks >= 60:
            print("Grade A")
        elif 45 < self.marks < 60:
            print("Grade B")
        else:
            print("Grade C")  
t = student(45,"Lalasa",55)
```

```python
t.display()
```

```python
t.grade()
```

```python
# calling instance method inside the constructor
class student:
    def __init__(self,rollno,name,marks):
        self.rollno = rollno
        self.name = name
        self.marks = marks
        self.grade()
    def display(self):
        return f" Rollno:{self.rollno}, name : {self.name} , marks : {self.marks}"
    def grade(self):
        if self.marks >= 60:
            print("Grade A")
        elif 45 < self.marks < 60:
            print("Grade B")
        else:
            print("Grade C")  
t = student(45,"Lalasa",55)
```

# Class Methods

- Works with class variables.
- First Parameter is cls.
- Created by using @classmethod decorater.

```python
# create a class where we can count the number of objects created.

class Test:
    object_count = 0
    def __init__(self):
        Test.object_count += 1
    @classmethod
    def show_count(cls):
        return f"The object count is {cls.object_count}"
```

```python
t1 = Test()
t1.show_count()
```

```python
# Currency Exchange

class CurrencyExchange:
    currency_rate = 91.00
    @classmethod
    def update_currency(cls):
        rate = float(input("enter the new rate"))
        cls.currency_rate = rate
    def convert(self,rupee_amount):
        return rupee_amount * CurrencyExchange.currency_rate
c1 = CurrencyExchange()
```

```python
c1.convert(1000)
```

```python
c1.update_currency()
```

```python
CurrencyExchange.currency_rate
```

- Behaves like regular function but wrapped inside a class.
